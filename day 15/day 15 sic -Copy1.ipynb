Hash Map



The % (modulus) operator is the bridge between a hash function and the hash table size.
It keeps the hash value inside valid array indexes.

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(size)]

    # Hash function    
    def hash_function(self, key):
        return key % self.size

    # Insert element   
    def insert(self, key):
        index = self.hash_function(key)
        self.table[index].append(key)
        print(f"Inserted {key} at index {index}")

    # Search element
    def search(self, key):
        index = self.hash_function(key)
        if key in self.table[index]:
            print(f"Element {key} found at index {index}")
            return True
        else:
            print(f"Element {key} not found")
            return False

    # Delete element
    def delete(self, key):
        index = self.hash_function(key)
        if key in self.table[index]:
            self.table[index].remove(key)
            print(f"Element {key} deleted")
        else:
            print(f"Element {key} not found for deletion")

    # Display hash table
    def display(self):
        for i in range(self.size):
            print(f"Index {i}: {self.table[i]}")


# ---- Testing the Hash Table ----
h = HashTable(10)

h.insert(10)
h.insert(20)
h.insert(15)
h.insert(7)
h.insert(27)
h.insert(37)

h.display()

h.search(15)
h.search(100)

h.delete(20)
h.delete(100)

h.display()
```
