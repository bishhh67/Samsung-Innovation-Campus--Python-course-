#GLOBAL


```python
x=11
def func():
    global x
    x +=2
    print(x)

print(x)
func()
print(x)
```

```python
x=11
def func():
    x=1

    def func2():
        global x
        x=2
        print(x)
    print(x)
    func2()
    print(x)

print(x)
func()
print(x)
```

```python
#Non Local
```

```python
x=11
def func():
    x=1

    def func2():
        nonlocal x
        x=2
        print(x)
    print(x)
    func2()
    print(x)


func()
```

```python

def counter(x=1):
                # x stays as memory , so we can access it later also when counter() is closed
    def plus():
        nonlocal x
        x+=1       # to do any operation in outer var , nonlocal must be used , otherwise a copy is created
        return x
        
    return plus           # return function ( plus )  , not value after calling function ( plus()) , 

count = counter(10)

print(count())

print(count())
print(count())
print(count())
```
