```python

table = {
    1000: "M", 900: "CM", 500: "D", 400: "CD",
    100: "C", 90: "XC", 50: "L", 40: "XL",
    10: "X", 9: "IX", 5: "V", 4: "IV",
    1: "I"
}


values = sorted(table.keys(), reverse=True)

def arabic_to_roman(number):
    roman = ""

    for value in values:
        while number >= value:
            roman += table[value]
            number -= value

    return roman


 
numbers = [369, 80, 29, 155, 14, 492, 348, 301, 469, 499]

for num in numbers:
    print(num, "=", arabic_to_roman(num))
```

```python
class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]

    def hash(self, key):
        if isinstance(key, int):
            return key % self.size
        elif not isinstance(key, str):
            raise ValueError("Key can only be str or int")
        ord_sum = sum(map(ord, key))
        return ord_sum % self.size
    
    def put(self, key, value):
        loc = self.hash(key)
        to_put = self.table[loc]
        for i, (k, v) in enumerate(to_put):
            if k == key:
                to_put[i] = (key, value)
                break
        else:
            # runs when break is not called, no break...
            to_put.append((key, value))
        # Equvalent code
        # flag = 0
        # for i, k, v in enumerate(to_put):
        #     if k == key:
        #         to_put[i] = (key, value)
        #         flag = 1
        #         break
        # if not flag:
        #     to_put.append((key, value))
    def get(self, key):

        loc = self.hash(key)
        to_get = self.table[loc]
        for k,v in to_get:
            if k == key:
                # early return
                return v
        raise KeyError("Key not found")

h = HashTable(20)
assert h.table == [[] for _ in range(20)]
assert h.hash(10) == 10
h = HashTable()
assert h.table == [[] for _ in range(10)]
assert h.hash("a") == 7
assert h.hash("A") == 5
h.put("a", "b")
assert ("a", "b") in h.table[7]
assert h.get("a") == "b"

try:
    h.get("b")
    assert False
except KeyError:
    assert True
h.put("a", "c")
assert h.get("a") == "c"

```
