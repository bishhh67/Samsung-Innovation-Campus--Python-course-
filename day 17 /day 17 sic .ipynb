```python
import turtle
import random
import math

screen = turtle.Screen()
screen.bgcolor("black")

t = turtle.Turtle()
t.speed(0)
t.width(2)

colors = ["red", "yellow", "cyan", "magenta", "lime", "orange", "white"]

turtle.tracer(0)

for i in range(120):
    t.color(colors[i % len(colors)])
    t.forward(i * 2)
    t.left(59)
    t.circle(80)

    turtle.update()

t.hideturtle()
turtle.done()
```
