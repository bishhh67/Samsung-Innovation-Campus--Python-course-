```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return Point(self.x+other.x,self.y+other.y)

    def __mul__(self, other):
        # should return a new Point
        return Point(self.x * other.x, self.y * other.y)

        

#    def __eq__(self,other):
 #       if self.x==other.x and self.y ==other.y:
  #          return "equal points"
   #     else:
    #        return "not equal"

    def __eq__(self,other):
        return self.x==other.x and self.y ==other.y      # this also returns boolean values 
        

            
    def __str__(self):                           # print() obj gardaa call hune function
        # can be anything inside this
        return f"point ({self.x},{self.y})"
    def __repr__(self):                           # simply objj ko naam lekhdaa run hune ftn , (without print)
        #can be anything , but prefered to have a value you can copy to make a new class
        return f"point is(x={self.x},y={self.y})"


a = Point(2,3)
b= Point(3,4)

print(a*b)   # sim to a.__mul__(b) ..... operator overloading in python

print(a+b)

a==b
    
```
