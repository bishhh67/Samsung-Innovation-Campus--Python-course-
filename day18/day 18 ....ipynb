```python
# quick sort


def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[0]
    left = []
    right = []

    for i in arr[1:]:
        if i < pivot:
            left.append(i)
        else:
            right.append(i)

    return quick_sort(left) + [pivot] + quick_sort(right)

numbers =[3,7,8,9,4,1,7,6,5]
sortedd = quick_sort(numbers)

print(numbers)
print(sortedd)
```
