```python
# ex 6

# qn1

def fact(n):
    if n==0:
        return 1

    else:
        return n* fact(n-1)

y=fact(5)
y
```

```python
# qn 2
def sum_(n):
    if n==0:
        return 0
    else:
        return n + sum_(n-1)

z= sum_(9)
z
```

```python
# qn 4

def power_(base,exp):
    if exp==0:
        return 1

    else:
        return base * power_(base,exp-1)

x=power_(2,5)
x
```

```python
# qn 5

def fib(n):
    if n==0:
        return 0
    elif n==2 or n==1:
        return 1

    else:
        return fib(n-2) + fib(n-1)

z = fib(7) 
z
        
```
