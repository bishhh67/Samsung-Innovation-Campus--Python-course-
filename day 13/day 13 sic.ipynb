```python
try:
    a=2
    b=3
    print(a+b)
except Exception as e:
    print("some error:",e)

else:
    print("no error came")

finally:
    print("i will always run")
    
```

```python
def func(div):
    try:
        print(" in try block")
        a=10
        print(a/div)
        return "no error cameeeee"
        
    except Exception as e:
        print("some error:",e)
        return "error came"
    
    else:
        print("no error came")
    
    finally:
        print("i will always run")

func(10)
print()
func(0)
```

```python
# finally block always run irrespective of you return something early or not
# in every case finally block will run
# but else block will not run incase of early return
```

```python
import re

re.search("name","my name is name bishh name")

# finds only the first time it occurs from left to right
```

```python
re.search("\d","my name is name 123 bishh name 345")

# \d    finds the first number in the string
```

```python
re.search("\d+","my name is name 123 bishh name 345")

# finds then first series of number
```

```python
re.findall("\d","my name is name 123 bishh name 345")

# finds all nos
```

```python
re.findall("\d+","my name is name 123 bishh name 345")

# finds all series of  nos
```

```python
s= """
My name is bish
my age is 87
my contact is 9861023453 """

re.search(r"\d{10}",s)   # finds the 10 unit long number
```

```python
s2= """ age is 45
contact = 984 234 4566 """

x=re.search(r"\d{3} \d{3} \d{4}",s2)
x
```

```python
s2= """ age is 45
contact = +1 984 234 4566 """

x=re.search(r"(\+1)? \d{3} \d{3} \d{4}",s2)
x
```

```python
s2= """ age is 45
contact = +1 (984) 234 4566 """

x=re.search(r"(\+1)? \(?\d{3}\)? \d{3} \d{4}",s2)
x
```

```python
s2= """ age is 45
contact = 984-234-4566 """

x=re.search(r"(\+1)?\(?\d{3}\)?[ -] \d{3}[ -]\d{4}",s2)
x
```
