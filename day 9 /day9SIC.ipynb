# LCM IS THE LOWEST NUMBER THAT CAN BE DIVIDED BY BOTH A AND B

```python
# qn 5

def lcm_(a, b):
    i = 2
    while True:
        if i % a == 0 and i % b == 0:   # FIxed: now here the lcm_num is being by a and b
            break
        i += 1
    return i

a, b = map(int, input("enter 2 nums: ").split())

l = lcm_(a, b)

print(l)
```
