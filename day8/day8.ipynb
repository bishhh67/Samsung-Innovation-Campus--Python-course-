#SERIALIZATION

Serialization in Python means converting a Python object into a format that can be saved or sent, and then recreating it back later.

Think of it like packing and unpacking data.

//////Why do we need serialization?

Python objects: live in RAM , disappear when the program ends

Serialization lets you:save data to a file,send data over a network,store data in a database,share data between programs
