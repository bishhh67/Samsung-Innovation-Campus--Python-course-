# create new list of higher or lower by checking next element

```python
# create new list of higher or lower by checking next element

l = [1,2,3,4,5,1,2,3]
neww=[]
for i in range(len(l)-1):
    for j in range(i+1 ,range(len(l))-1):
      if l[j]>l[i]:
        neww.append("higher")
      else:
        neww.append("lower")
print(neww)


```
